```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
res = 'https://www.macrotrends.net/countries/ETH/ethiopia/population'
```

```python
r = requests.get(res)
```

```python
print(r.ok)
print(r.status_code)
```

```python
soup = BeautifulSoup(r.content, 'html')
```

```python
print(soup.prettify())
#print(soup)
```

```python
table = soup.find_all('table')
```

```python
table[1]
```

```python
df = pd.read_html(str(table[1]))
```
